```python
import numpy as np
import matplotlib.pyplot as plt

def triplet_loss(anchor, positive, negative, margin=1.0):
    pos_dist = np.sum((anchor - positive) ** 2)
    neg_dist = np.sum((anchor - negative) ** 2)
    loss = np.maximum(0, pos_dist - neg_dist + margin)
    return loss

# Example for data points
anchor = np.array([1.0, 2.0])
positive = np.array([1.0, 3.0])
negative = np.array([3.0, 2.0])
margin = 1.0

loss_value = triplet_loss(anchor, positive, negative, margin) # loss calculate
print(f"Triplet Loss: {loss_value}")

# Visualize
plt.figure(figsize=(6, 6))
plt.scatter(*anchor, color='blue', label='Anchor (A)', s=100)
plt.scatter(*positive, color='green', label='Positive (P)', s=100)
plt.scatter(*negative, color='red', label='Negative (N)', s=100)

plt.plot([anchor[0], positive[0]], [anchor[1], positive[1]], 'g--')
plt.plot([anchor[0], negative[0]], [anchor[1], negative[1]], 'r--')

plt.text(anchor[0], anchor[1], 'A', fontsize=12, ha='right')
plt.text(positive[0], positive[1], 'P', fontsize=12, ha='left')
plt.text(negative[0], negative[1], 'N', fontsize=12, ha='right')
plt.legend()
plt.title(f'Triplet Loss Visualization\nLoss Value: {loss_value:.2f}')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.grid(True)
plt.axis('equal')
plt.show()
```

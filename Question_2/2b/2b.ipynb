```python
import numpy as np
import matplotlib.pyplot as plt

def extended_triplet_loss(anchor, positives, negatives, alpha=0.2):
    loss = 0.0
    
    for positive in positives:
        for negative in negatives:
            pos_dist = np.sum((anchor - positive) ** 2)
            neg_dist = np.sum((anchor - negative) ** 2)
            loss += np.maximum(0, pos_dist - neg_dist + alpha)
    
    return loss

def plot_triplet_loss(anchor, positives, negatives):
    plt.figure(figsize=(8, 8))
    
    # Plot anchor
    plt.scatter(anchor[0], anchor[1], color='blue', label='Anchor', s=100, edgecolor='black')
    
    # Plot positives
    for i, positive in enumerate(positives):
        plt.scatter(positive[0], positive[1], color='green', label=f'Positive {i+1}', s=100, edgecolor='black')
        plt.plot([anchor[0], positive[0]], [anchor[1], positive[1]], 'g--')
    
    # Plot negatives
    for i, negative in enumerate(negatives):
        plt.scatter(negative[0], negative[1], color='red', label=f'Negative {i+1}', s=100, edgecolor='black')
        plt.plot([anchor[0], negative[0]], [anchor[1], negative[1]], 'r--')
    
    plt.legend(loc='best')
    plt.title('Visualization of Extended Triplet Loss')
    plt.xlabel('Dimension 1')
    plt.ylabel('Dimension 2')
    plt.grid(True)
    plt.show()

anchor = np.array([1.0, 2.0])
positives = [np.array([1.1, 2.1]), np.array([0.9, 1.9])]
negatives = [
    np.array([3.0, 4.0]),
    np.array([4.0, 5.0]),
    np.array([5.0, 6.0]),
    np.array([6.0, 7.0]),
    np.array([7.0, 8.0])
]

loss = extended_triplet_loss(anchor, positives, negatives)
print("Extended Triplet Loss:", loss)

# Visualize
plot_triplet_loss(anchor, positives, negatives)
```
